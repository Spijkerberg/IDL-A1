```python
import numpy as np
import pandas as pd 

csv = pd.read_csv(r"C:\Users\timvd\Documents\Uni 2023-2024\IDL\IDL-A1\individual_solns\IDL_A1_convolution_with_BN.csv")
kwargs = np.array(csv['Kwargs'])
etas = [element[-6:-1] for element in kwargs]
csv = csv.drop('Kwargs',axis='columns')
csv['Learning rate'] = etas
csv = csv.rename(columns={"Name":"Optimizer","Batch_size":"Batch size"})
cols = csv.columns.tolist()
cols = [cols[0],cols[-1],*cols[1:-1]]
csv = csv[cols]
csv = csv.round({'Accuracy':3})
csv.sort_values('Accuracy',inplace=True,ascending=False,ignore_index=True)
csv.to_csv(r"C:\Users\timvd\Documents\Uni 2023-2024\IDL\IDL-A1\individual_solns\IDL_A1_convolution_with_BN_forLatex.csv",index=False)
csv
```
